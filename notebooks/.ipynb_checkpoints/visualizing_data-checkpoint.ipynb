```python
# Import Dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the spiral data from the CSV file
file_path = '../data/spiral_data.csv'  # Update the path if necessary
spiral_data = pd.read_csv(file_path)

# Extract x and y coordinates
x = spiral_data['x']
y = spiral_data['y']

# Plot the spiral
plt.figure(figsize=(10, 10))
plt.scatter(x, y, color='blue', alpha=0.7, s=5)
plt.axis('equal')
plt.title("Dados originais")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
import numpy as np

# Parameters for DBSCAN
eps = 0.7  # Maximum distance between two samples to be considered neighbors
min_samples = 3  # Minimum number of points in a neighborhood to form a cluster

# Stack x and y into a single array for clustering
coordinates = np.column_stack((x, y))

# Apply DBSCAN
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
labels = dbscan.fit_predict(coordinates)

# Plot the clusters
plt.figure(figsize=(10, 10))
unique_labels = set(labels)
for label in unique_labels:
    if label == -1:  # Noise points
        color = 'gray'
        alpha = 0.3
    else:
        color = None  # Let matplotlib auto-assign colors to clusters
        alpha = 0.8
    cluster_points = coordinates[labels == label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {label}", alpha=alpha)
print('NÃºmero de Clusters:', len(unique_labels))
plt.axis('equal')
plt.title("DBSCAN Clustering of Spiral Data")
plt.show()
```

```python
# Kernel Density
from sklearn.neighbors import KernelDensity
from matplotlib.colors import Normalize

# Fit KDE
kde = KernelDensity(kernel='gaussian', bandwidth=0.05).fit(coordinates)

# Evaluate KDE on a grid
x_grid = np.linspace(x.min(), x.max(), 200)
y_grid = np.linspace(y.min(), y.max(), 200)
X_grid, Y_grid = np.meshgrid(x_grid, y_grid)
grid_points = np.column_stack([X_grid.ravel(), Y_grid.ravel()])

# Calculate density values
log_density = kde.score_samples(grid_points)
density = np.exp(log_density).reshape(X_grid.shape)

# Plot KDE density
plt.figure(figsize=(6, 6))
plt.pcolormesh(X_grid, Y_grid, density, shading='auto', cmap='viridis', norm=Normalize(vmin=0, vmax=density.max()))
plt.colorbar(label="Density")
plt.scatter(x, y, color='blue', alpha=0.5, s=10, label="Data Points")
plt.title("KDE Density Estimation of Spiral Data")
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA

def kmeans_plus_plus_initialization(data, k):
    """
    Implements the K-means++ initialization algorithm to find k centroids.

    Parameters:
        data (numpy.ndarray): A 2D array of shape (n_samples, n_features) where each row is a data point.
        k (int): The number of centroids to initialize.

    Returns:
        centroids (numpy.ndarray): A 2D array of shape (k, n_features) containing the initialized centroids.
    """
    n_samples, n_features = data.shape
    
    # Step 1: Randomly choose the first centroid from the data points
    centroids = [data[np.random.choice(n_samples)]]
    
    # Step 2: Select the remaining k-1 centroids
    for _ in range(1, k):
        # Compute the squared distances from each point to the nearest centroid
        distances = np.array([
            min(np.sum((point - centroid)**2) for centroid in centroids) 
            for point in data
        ])
        
        # Probability of choosing a point is proportional to the squared distance
        probabilities = distances / np.sum(distances)
        
        # Choose the next centroid based on the computed probabilities
        next_centroid = data[np.random.choice(n_samples, p=probabilities)]
        centroids.append(next_centroid)
    
    return np.array(centroids)


def distance_point_to_segment(point, segment):
    # Descompactar coordenadas
    px, py = point
    segment_start, segment_end = segment
    x1, y1 = segment_start
    x2, y2 = segment_end

    # Vetores
    AP = np.array([px - x1, py - y1])
    AB = np.array([x2 - x1, y2 - y1])
    
    # Comprimento ao quadrado do segmento
    AB_length_squared = AB.dot(AB)
    
    if AB_length_squared == 0:
        # A e B sÃ£o o mesmo ponto
        return np.linalg.norm(AP)
    
    # ProjeÃ§Ã£o escalar
    t = AP.dot(AB) / AB_length_squared

    # Restringir t ao intervalo [0, 1]
    t = max(0, min(1, t))
    
    # Ponto projetado no segmento
    projection = np.array([x1, y1]) + t * AB
    
    # DistÃ¢ncia euclidiana entre P e a projeÃ§Ã£o
    distance = np.linalg.norm(np.array([px, py]) - projection)
    
    return distance

def init_segments_as_points(data, k): 
    centroids = kmeans_plus_plus_initialization(data, k)
    # Nesse caso, retornamos segmentos com comeÃ§o e fim iguais, ou seja, com comprimento = 0 
    # jÃ¡ que nÃ£o sabemos nada sobre seu agrupamento de dados
    segments = np.array([np.array([point, point]) for point in centroids])
    return segments

# 2. AtribuiÃ§Ã£o aos segmentos mais prÃ³ximos
def assign_points_to_segments(data, segments):
    # Retorna as regiÃµes de Voronoi baseadas nos segmentos
    assignments = []
    for point in data:
        distances = [distance_point_to_segment(point, segment) for segment in segments]
        assignments.append(np.argmin(distances))
    return assignments

# 3. Atualizar segmentos
def update_segments(data, assignments, k):
    # TODO: Calcular o tamanho dos segmentos usando 3/2 vezes o desvio padrÃ£o
    segments = []
    for i in range(k):
        points_in_region = data[np.array(assignments) == i]
        if len(points_in_region) > 0:
            pca = PCA(n_components=1)
            pca.fit(points_in_region)
            direction = pca.components_[0]
            mean = cluster_points.mean(axis=0)
            projected_centroid = mean + np.dot(mean - cluster_points.mean(axis=0), direction) * direction
            new_centroids.append(projected_centroid)        
    return segments

# FunÃ§Ã£o principal
def k_segments_algorithm(data, k=5, n_iterations = 1):
    segments = init_segments_as_points(data, k)
    print(segments)
    for i in range(n_iterations):
        print('IteraÃ§Ã£o: ', i)
        voronoi_regions = assign_points_to_segments(data, segments)
        segments = update_segments(data, voronoi_regions, k)
        print('segments: ', segments)
        return segments
        
initial_clusters = k_segments_algorithm(data, 5, 1)

plt.figure(figsize=(10, 8))
plt.scatter(data[:, 0], data[:, 1], color=initial_clusters, label='Data Points', alpha=0.5)

plt.legend()
plt.title('Principal Curve using Verbeek Algorithm')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.interpolate import splprep, splev
from IPython.display import HTML

# Generate synthetic dataset
np.random.seed(42)
t = np.linspace(0, 4 * np.pi, 300)
x = t * np.cos(t) + np.random.normal(0, 0.1, len(t))
y = t * np.sin(t) + np.random.normal(0, 0.1, len(t))
data = np.vstack([x, y]).T

# Parameters for the Verbeek algorithm
num_segments = 10
max_iterations = 10

# Initialize centroids using k-means
kmeans = KMeans(n_clusters=num_segments, random_state=42)
kmeans.fit(data)
centroids = kmeans.cluster_centers_

centroids_history = []

# Verbeek Algorithm Refinement
for _ in range(max_iterations):
    distances = np.linalg.norm(data[:, None, :] - centroids[None, :, :], axis=2)
    nearest_centroid = np.argmin(distances, axis=1)
    new_centroids = []

    for i in range(num_segments):
        cluster_points = data[nearest_centroid == i]
        if len(cluster_points) > 0:
            pca = PCA(n_components=1)
            pca.fit(cluster_points)
            direction = pca.components_[0]
            mean = cluster_points.mean(axis=0)
            projected_centroid = mean + np.dot(mean - cluster_points.mean(axis=0), direction) * direction
            new_centroids.append(projected_centroid)
        else:
            new_centroids.append(centroids[i])
    centroids = np.array(new_centroids)
    centroids_history.append(centroids.copy())

# Final smoothing
tck, _ = splprep(centroids.T, s=2)
final_curve = np.array(splev(np.linspace(0, 1, 1000), tck)).T

# Animation
fig, ax = plt.subplots(figsize=(8, 6))

def update(frame):
    ax.clear()
    centroids = centroids_history[frame]
    ax.scatter(data[:, 0], data[:, 1], alpha=0.5, label='Data Points')
    ax.scatter(centroids[:, 0], centroids[:, 1], color='red', label='Centroids')
    ax.legend()
    ax.set_title(f"Iteration {frame + 1}")

ani = FuncAnimation(fig, update, frames=len(centroids_history), repeat=False)
HTML(ani.to_jshtml())
```
