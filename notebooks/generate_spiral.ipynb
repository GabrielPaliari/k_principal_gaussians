```python
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral(n_points, noise=0.05):
    """
    Generate a single spiral dataset with noise.

    Parameters:
    - n_points (int): Number of points in the spiral.
    - noise (float): Amount of noise to add to the points.

    Returns:
    - X (numpy.ndarray): Array of x-coordinates.
    - Y (numpy.ndarray): Array of y-coordinates.
    """
    theta = np.linspace(0, 4 * np.pi, n_points)  # Angle for the spiral
    r = np.linspace(0, 5, n_points)  # Radius increases linearly
    x = r * np.cos(theta)  # x-coordinates
    y = r * np.sin(theta)  # y-coordinates

    # Add Gaussian noise
    x += np.random.normal(0, noise, n_points)
    y += np.random.normal(0, noise, n_points)
    
    return x, y

# Generate spiral data
n_points = 400
noise_level = 0.1
x, y = generate_spiral(n_points, noise_level)

# Plot the spiral
plt.figure(figsize=(10, 10))
plt.scatter(x, y, color='blue', alpha=0.7)
plt.axis('equal')
plt.title("Noisy Single Spiral")
plt.show()
```

```python
import os
import pandas as pd

# Create the data folder if it doesn't exist
data_folder = './data'
os.makedirs(data_folder, exist_ok=True)

# Combine x and y into a pandas DataFrame
spiral_data = pd.DataFrame({
    'x': x,
    'y': y
})

# Save the data to a CSV file
file_path = os.path.join(data_folder, 'spiral_data.csv')
spiral_data.to_csv(file_path, index=False)

print(f"Spiral data saved to {file_path}")
```
